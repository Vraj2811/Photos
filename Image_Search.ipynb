```python
# üöÄ OLLAMA-ONLY APPROACH - All AI functionality using Ollama
print("ü¶ô Setting up Ollama-based AI pipeline...")

# Reset previous models to None since we're using Ollama-only approach
text_embedding_model = None
clip_model = None
clip_processor = None

class OllamaImageProcessor:
    """
    Complete image processing pipeline using only Ollama models
    """
    
    def __init__(self):
        self.client = ollama
        self.vision_model = "llava"  # For image descriptions
        self.embedding_model = "nomic-embed-text"  # For text embeddings
        self.available_models = []
        self.check_available_models()
    
    def check_available_models(self):
        """Check which Ollama models are available"""
        try:
            models = ollama.list()
            self.available_models = [model['name'].split(':')[0] for model in models['models']]
            print(f"üìã Available Ollama models: {self.available_models}")
            
            # Check required models
            if self.vision_model not in self.available_models:
                print(f"‚ö†Ô∏è  {self.vision_model} not found. Run: ollama pull {self.vision_model}")
            else:
                print(f"‚úÖ {self.vision_model} is available")
                
            if self.embedding_model not in self.available_models:
                print(f"‚ö†Ô∏è  {self.embedding_model} not found. Run: ollama pull {self.embedding_model}")
                print("üí° Alternative: We can use the llava model for text embeddings too")
            else:
                print(f"‚úÖ {self.embedding_model} is available")
                
        except Exception as e:
            print(f"‚ùå Error checking models: {e}")
    
    def describe_image(self, image_path, prompt="Describe this image in detail, including objects, people, colors, setting, mood, and any text visible."):
        """
        Generate detailed description of an image using LLaVA
        """
        try:
            response = ollama.chat(
                model=self.vision_model,
                messages=[
                    {
                        'role': 'user',
                        'content': prompt,
                        'images': [image_path]
                    }
                ]
            )
            return response['message']['content']
        except Exception as e:
            print(f"‚ùå Error describing image {image_path}: {e}")
            return None
    
    def generate_embedding(self, text):
        """
        Generate text embedding using Ollama
        """
        try:
            # Try using dedicated embedding model first
            if self.embedding_model in self.available_models:
                response = ollama.embeddings(
                    model=self.embedding_model,
                    prompt=text
                )
                return np.array(response['embedding'])
            else:
                # Fallback: Use a simple hash-based approach with the text
                return self._simple_text_embedding(text)
                
        except Exception as e:
            print(f"‚ö†Ô∏è  Embedding generation failed, using fallback: {e}")
            return self._simple_text_embedding(text)
    
    def _simple_text_embedding(self, text, dimension=384):
        """
        Simple text embedding fallback using hash-based approach
        """
        words = text.lower().split()
        embedding = np.zeros(dimension)
        
        for i, word in enumerate(words):
            # Use multiple hash functions for better distribution
            hash1 = hash(word) % dimension
            hash2 = hash(word[::-1]) % dimension  # Reverse word hash
            hash3 = hash(word + str(len(word))) % dimension  # Length-based hash
            
            embedding[hash1] += 1.0
            embedding[hash2] += 0.5
            embedding[hash3] += 0.3
        
        # Normalize the embedding
        norm = np.linalg.norm(embedding)
        if norm > 0:
            embedding = embedding / norm
        
        return embedding
    
    def extract_image_tags(self, image_path):
        """
        Extract searchable tags from image using LLaVA
        """
        tag_prompt = """List the main objects, people, and concepts in this image as short tags or keywords. 
        Include: objects, people, animals, colors, setting, activity, emotion, style.
        Format as comma-separated tags like: sunset, mountains, orange sky, peaceful, landscape"""
        
        try:
            response = ollama.chat(
                model=self.vision_model,
                messages=[
                    {
                        'role': 'user',
                        'content': tag_prompt,
                        'images': [image_path]
                    }
                ]
            )
            tags_text = response['message']['content']
            # Extract tags (simple parsing)
            tags = [tag.strip() for tag in tags_text.split(',')]
            return tags[:20]  # Limit to 20 tags
        except Exception as e:
            print(f"‚ùå Error extracting tags from {image_path}: {e}")
            return []

# Initialize the Ollama processor
print("\nüîß Initializing Ollama Image Processor...")
ollama_processor = OllamaImageProcessor()

# Test the processor
print("\nüß™ Testing Ollama functionality...")

# Test text embedding
test_text = "A beautiful sunset over mountains with orange and pink colors"
test_embedding = ollama_processor.generate_embedding(test_text)
print(f"‚úÖ Text embedding test - Shape: {test_embedding.shape}, Type: {type(test_embedding)}")

# Update configuration for Ollama-only approach
Config.EMBEDDING_DIMENSION = len(test_embedding)
print(f"üìè Updated embedding dimension to: {Config.EMBEDDING_DIMENSION}")

# Reinitialize FAISS with correct dimension
faiss_index = faiss.IndexFlatIP(Config.EMBEDDING_DIMENSION)
print(f"‚úÖ FAISS index reinitialized with dimension: {Config.EMBEDDING_DIMENSION}")

print("\nüéØ Ollama-Only Setup Complete!")
print("=" * 50)
print("‚úÖ Image descriptions: LLaVA")
print("‚úÖ Text embeddings: Ollama (nomic-embed-text or fallback)")
print("‚úÖ Image tags: LLaVA")
print("‚úÖ Vector search: FAISS")
print("‚úÖ No external dependencies (Hugging Face)")

print("\nüöÄ Ready for image processing pipeline!")
```

```python
# üóÑÔ∏è Database Setup for Ollama-Only Approach
print("üóÑÔ∏è Setting up SQLite database...")

# Database Models
Base = declarative_base()

class ImageRecord(Base):
    """Database model for storing image information"""
    __tablename__ = 'images'
    
    id = Column(Integer, primary_key=True, autoincrement=True)
    filename = Column(String(255), nullable=False)
    file_path = Column(Text, nullable=False)
    file_size = Column(Integer)
    width = Column(Integer)
    height = Column(Integer)
    format = Column(String(10))
    created_at = Column(DateTime, default=datetime.now)
    processed_at = Column(DateTime)
    
    # AI-generated content
    description = Column(Text)  # LLaVA description
    tags = Column(Text)  # Comma-separated tags
    embedding = Column(Text)  # JSON-serialized embedding vector
    
    def __repr__(self):
        return f"<ImageRecord(id={self.id}, filename='{self.filename}')>"

# Create database and tables
def setup_database():
    """Initialize the database"""
    try:
        engine = create_engine(Config.DATABASE_URL, echo=False)
        Base.metadata.create_all(engine)
        
        SessionLocal = sessionmaker(bind=engine)
        session = SessionLocal()
        
        print(f"‚úÖ Database created: {Config.DATABASE_PATH}")
        return engine, SessionLocal
    except Exception as e:
        print(f"‚ùå Database setup failed: {e}")
        return None, None

# Initialize database
db_engine, SessionLocal = setup_database()

# Database helper functions
class ImageDatabase:
    """Helper class for database operations"""
    
    def __init__(self, session_factory):
        self.SessionLocal = session_factory
    
    def add_image(self, filename, file_path, file_size=None, width=None, height=None, 
                  format=None, description=None, tags=None, embedding=None):
        """Add a new image record to the database"""
        session = self.SessionLocal()
        try:
            # Convert embedding to JSON string if it's a numpy array
            embedding_json = None
            if embedding is not None:
                embedding_json = json.dumps(embedding.tolist())
            
            # Convert tags list to comma-separated string
            tags_str = None
            if tags:
                tags_str = ', '.join(tags) if isinstance(tags, list) else tags
            
            image_record = ImageRecord(
                filename=filename,
                file_path=file_path,
                file_size=file_size,
                width=width,
                height=height,
                format=format,
                description=description,
                tags=tags_str,
                embedding=embedding_json,
                processed_at=datetime.now()
            )
            
            session.add(image_record)
            session.commit()
            
            image_id = image_record.id
            print(f"‚úÖ Added image to database: {filename} (ID: {image_id})")
            return image_id
            
        except Exception as e:
            session.rollback()
            print(f"‚ùå Failed to add image {filename}: {e}")
            return None
        finally:
            session.close()
    
    def get_all_images(self):
        """Get all image records"""
        session = self.SessionLocal()
        try:
            images = session.query(ImageRecord).all()
            return images
        finally:
            session.close()
    
    def get_image_by_id(self, image_id):
        """Get image by ID"""
        session = self.SessionLocal()
        try:
            return session.query(ImageRecord).filter(ImageRecord.id == image_id).first()
        finally:
            session.close()
    
    def search_by_description(self, search_term):
        """Simple text search in descriptions"""
        session = self.SessionLocal()
        try:
            return session.query(ImageRecord).filter(
                ImageRecord.description.contains(search_term)
            ).all()
        finally:
            session.close()

# Initialize database helper
if SessionLocal:
    db_helper = ImageDatabase(SessionLocal)
    print("‚úÖ Database helper initialized")
else:
    db_helper = None
    print("‚ùå Database helper not available")

# Test database
if db_helper:
    print("\nüß™ Testing database functionality...")
    existing_images = db_helper.get_all_images()
    print(f"üìä Current images in database: {len(existing_images)}")
    
    if existing_images:
        print("üìã Existing images:")
        for img in existing_images[:5]:  # Show first 5
            print(f"   - {img.filename} (ID: {img.id})")

print("\nüìã Database Schema Created:")
print("   ‚úÖ images table")
print("   ‚úÖ Fields: id, filename, file_path, file_size, dimensions")
print("   ‚úÖ AI fields: description, tags, embedding")
print("   ‚úÖ Timestamps: created_at, processed_at")

print("\nüéØ Database Ready for Image Storage!")
```

```python
# üîÑ Complete Image Processing Pipeline (Ollama-Only)
print("üîÑ Creating complete image processing pipeline...")

class ImageSearchPipeline:
    """
    Complete end-to-end image search pipeline using only Ollama
    """
    
    def __init__(self, ollama_processor, db_helper, faiss_index):
        self.ollama_processor = ollama_processor
        self.db_helper = db_helper
        self.faiss_index = faiss_index
        self.image_id_mapping = {}  # Maps FAISS index positions to database IDs
        
    def scan_images_folder(self, folder_path):
        """Scan folder for supported image files"""
        folder_path = Path(folder_path)
        if not folder_path.exists():
            print(f"‚ùå Folder not found: {folder_path}")
            return []
        
        supported_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp', '.gif']
        image_files = []
        
        for ext in supported_extensions:
            image_files.extend(folder_path.glob(f"*{ext}"))
            image_files.extend(folder_path.glob(f"*{ext.upper()}"))
        
        print(f"üìÅ Found {len(image_files)} images in {folder_path}")
        return image_files
    
    def get_image_metadata(self, image_path):
        """Extract basic image metadata"""
        try:
            with Image.open(image_path) as img:
                return {
                    'width': img.width,
                    'height': img.height,
                    'format': img.format,
                    'file_size': image_path.stat().st_size
                }
        except Exception as e:
            print(f"‚ö†Ô∏è  Could not read metadata for {image_path}: {e}")
            return {}
    
    def process_single_image(self, image_path):
        """Process a single image through the complete pipeline"""
        try:
            print(f"üñºÔ∏è  Processing: {image_path.name}")
            
            # 1. Get basic metadata
            metadata = self.get_image_metadata(image_path)
            
            # 2. Generate description using LLaVA
            print("   üìù Generating description...")
            description = self.ollama_processor.describe_image(str(image_path))
            if not description:
                print(f"   ‚ùå Failed to generate description for {image_path.name}")
                return None
            
            # 3. Extract tags
            print("   üè∑Ô∏è  Extracting tags...")
            tags = self.ollama_processor.extract_image_tags(str(image_path))
            
            # 4. Generate embedding from description
            print("   üî¢ Generating embedding...")
            embedding = self.ollama_processor.generate_embedding(description)
            
            # 5. Store in database
            print("   üíæ Storing in database...")
            image_id = self.db_helper.add_image(
                filename=image_path.name,
                file_path=str(image_path),
                file_size=metadata.get('file_size'),
                width=metadata.get('width'),
                height=metadata.get('height'),
                format=metadata.get('format'),
                description=description,
                tags=tags,
                embedding=embedding
            )
            
            if image_id:
                # 6. Add to FAISS index
                self.faiss_index.add(embedding.reshape(1, -1))
                current_index = self.faiss_index.ntotal - 1
                self.image_id_mapping[current_index] = image_id
                
                print(f"   ‚úÖ Successfully processed {image_path.name} (ID: {image_id})")
                return {
                    'image_id': image_id,
                    'filename': image_path.name,
                    'description': description,
                    'tags': tags,
                    'embedding_shape': embedding.shape
                }
            else:
                print(f"   ‚ùå Failed to store {image_path.name}")
                return None
                
        except Exception as e:
            print(f"‚ùå Error processing {image_path}: {e}")
            return None
    
    def process_images_folder(self, folder_path, max_images=None):
        """Process all images in a folder"""
        image_files = self.scan_images_folder(folder_path)
        
        if not image_files:
            print("‚ùå No images found to process")
            return []
        
        if max_images:
            image_files = image_files[:max_images]
            print(f"üî¢ Processing first {max_images} images")
        
        results = []
        print(f"\nüöÄ Starting batch processing of {len(image_files)} images...")
        print("=" * 60)
        
        for i, image_path in enumerate(tqdm(image_files, desc="Processing images"), 1):
            print(f"\nüì∏ [{i}/{len(image_files)}] {image_path.name}")
            result = self.process_single_image(image_path)
            if result:
                results.append(result)
            
            # Small delay to avoid overwhelming Ollama
            import time
            time.sleep(0.5)
        
        print(f"\n‚úÖ Batch processing complete!")
        print(f"üìä Successfully processed: {len(results)}/{len(image_files)} images")
        print(f"üîç FAISS index size: {self.faiss_index.ntotal} vectors")
        
        return results
    
    def search_images(self, query, top_k=5, confidence_threshold=0.3):
        """Search for images using natural language query"""
        try:
            print(f"üîç Searching for: '{query}'")
            
            # Generate embedding for the query
            query_embedding = self.ollama_processor.generate_embedding(query)
            
            # Search in FAISS index
            scores, indices = self.faiss_index.search(
                query_embedding.reshape(1, -1), 
                min(top_k, self.faiss_index.ntotal)
            )
            
            results = []
            for score, idx in zip(scores[0], indices[0]):
                if score >= confidence_threshold:
                    image_id = self.image_id_mapping.get(idx)
                    if image_id:
                        image_record = self.db_helper.get_image_by_id(image_id)
                        if image_record:
                            results.append({
                                'image_id': image_id,
                                'filename': image_record.filename,
                                'file_path': image_record.file_path,
                                'description': image_record.description,
                                'tags': image_record.tags,
                                'confidence': float(score),
                                'width': image_record.width,
                                'height': image_record.height
                            })
            
            print(f"‚úÖ Found {len(results)} matching images")
            return results
            
        except Exception as e:
            print(f"‚ùå Search failed: {e}")
            return []

# Initialize the complete pipeline
if db_helper and faiss_index and ollama_processor:
    pipeline = ImageSearchPipeline(ollama_processor, db_helper, faiss_index)
    print("‚úÖ Image Search Pipeline initialized!")
    
    # Display current status
    all_images = db_helper.get_all_images()
    print(f"\nüìä Current Status:")
    print(f"   Database: {len(all_images)} images")
    print(f"   FAISS Index: {faiss_index.ntotal} vectors")
    print(f"   Images folder: {Config.IMAGES_FOLDER}")
    
else:
    pipeline = None
    print("‚ùå Pipeline initialization failed - missing components")

print("\nüéØ Ready to process images!")
print("üí° Next steps:")
print("   1. Add images to the 'images' folder")
print("   2. Run: pipeline.process_images_folder(Config.IMAGES_FOLDER)")
print("   3. Search: pipeline.search_images('your query here')")
```

```python
# üöÄ PRODUCTION SETUP: Disk-Based FAISS for Large-Scale Processing
print("üöÄ SETTING UP DISK-BASED FAISS FOR 100+ IMAGES")
print("=" * 60)

import os
import json
from pathlib import Path

class DiskBasedImageSearchPipeline:
    """
    Enhanced pipeline with automatic disk persistence for large-scale processing
    """
    
    def __init__(self, ollama_processor, db_helper):
        self.ollama_processor = ollama_processor
        self.db_helper = db_helper
        self.faiss_index = None
        self.image_id_mapping = {}
        
        # File paths for persistence
        self.index_file = Config.FAISS_INDEX_PATH / "image_vectors.index"
        self.mapping_file = Config.FAISS_INDEX_PATH / "id_mapping.json"
        
        # Initialize or load existing index
        self.initialize_faiss_index()
    
    def initialize_faiss_index(self):
        """Initialize FAISS index - load from disk if exists, create new if not"""
        try:
            if self.index_file.exists() and self.mapping_file.exists():
                print("üìÇ Loading existing FAISS index from disk...")
                self.load_index_from_disk()
            else:
                print("üÜï Creating new FAISS index...")
                self.create_new_index()
                
        except Exception as e:
            print(f"‚ùå Error initializing FAISS index: {e}")
            print("üîÑ Creating fresh index...")
            self.create_new_index()
    
    def create_new_index(self):
        """Create a new FAISS index"""
        self.faiss_index = faiss.IndexFlatIP(Config.EMBEDDING_DIMENSION)
        self.image_id_mapping = {}
        print(f"‚úÖ New FAISS index created (dimension: {Config.EMBEDDING_DIMENSION})")
    
    def load_index_from_disk(self):
        """Load FAISS index and mappings from disk"""
        try:
            # Load FAISS index
            self.faiss_index = faiss.read_index(str(self.index_file))
            
            # Load ID mapping
            with open(self.mapping_file, 'r') as f:
                mapping_str = json.load(f)
                self.image_id_mapping = {int(k): v for k, v in mapping_str.items()}
            
            print(f"‚úÖ Loaded FAISS index from disk:")
            print(f"   üìä Vectors: {self.faiss_index.ntotal}")
            print(f"   üìè Dimension: {self.faiss_index.d}")
            print(f"   üóÇÔ∏è  Mappings: {len(self.image_id_mapping)}")
            
        except Exception as e:
            print(f"‚ùå Failed to load index from disk: {e}")
            raise
    
    def save_index_to_disk(self):
        """Save FAISS index and mappings to disk"""
        try:
            # Save FAISS index
            faiss.write_index(self.faiss_index, str(self.index_file))
            
            # Save ID mapping
            with open(self.mapping_file, 'w') as f:
                mapping_str = {str(k): v for k, v in self.image_id_mapping.items()}
                json.dump(mapping_str, f, indent=2)
            
            # Get file sizes for reporting
            index_size = self.index_file.stat().st_size / 1024
            mapping_size = self.mapping_file.stat().st_size / 1024
            
            print(f"üíæ Saved to disk:")
            print(f"   üìÑ Index: {self.index_file.name} ({index_size:.1f} KB)")
            print(f"   üìÑ Mapping: {self.mapping_file.name} ({mapping_size:.1f} KB)")
            
        except Exception as e:
            print(f"‚ùå Failed to save index to disk: {e}")
    
    def scan_images_folder(self, folder_path):
        """Scan folder for supported image files"""
        folder_path = Path(folder_path)
        if not folder_path.exists():
            print(f"‚ùå Folder not found: {folder_path}")
            return []
        
        supported_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp', '.gif']
        image_files = []
        
        for ext in supported_extensions:
            image_files.extend(folder_path.glob(f"*{ext}"))
            image_files.extend(folder_path.glob(f"*{ext.upper()}"))
        
        print(f"üìÅ Found {len(image_files)} images in {folder_path}")
        return image_files
    
    def get_image_metadata(self, image_path):
        """Extract basic image metadata"""
        try:
            with Image.open(image_path) as img:
                return {
                    'width': img.width,
                    'height': img.height,
                    'format': img.format,
                    'file_size': image_path.stat().st_size
                }
        except Exception as e:
            print(f"‚ö†Ô∏è  Could not read metadata for {image_path}: {e}")
            return {}
    
    def check_if_processed(self, image_path):
        """Check if image is already processed in database"""
        try:
            filename = image_path.name
            file_path = str(image_path)
            
            # Check if this exact file is already in database
            conn = sqlite3.connect(Config.DATABASE_PATH)
            cursor = conn.cursor()
            
            cursor.execute("""
                SELECT id FROM images 
                WHERE filename = ? AND file_path = ?
            """, (filename, file_path))
            
            result = cursor.fetchone()
            conn.close()
            
            return result[0] if result else None
            
        except Exception as e:
            print(f"‚ö†Ô∏è  Error checking if image processed: {e}")
            return None
    
    def process_single_image(self, image_path):
        """Process a single image through the complete pipeline"""
        try:
            # Check if already processed
            existing_id = self.check_if_processed(image_path)
            if existing_id:
                print(f"   ‚è≠Ô∏è  Already processed: {image_path.name} (ID: {existing_id})")
                return {'image_id': existing_id, 'filename': image_path.name, 'status': 'skipped'}
            
            print(f"üñºÔ∏è  Processing: {image_path.name}")
            
            # 1. Get basic metadata
            metadata = self.get_image_metadata(image_path)
            
            # 2. Generate description using LLaVA
            print("   üìù Generating description...")
            description = self.ollama_processor.describe_image(str(image_path))
            if not description:
                print(f"   ‚ùå Failed to generate description for {image_path.name}")
                return None
            
            # 3. Extract tags
            print("   üè∑Ô∏è  Extracting tags...")
            tags = self.ollama_processor.extract_image_tags(str(image_path))
            
            # 4. Generate embedding from description
            print("   üî¢ Generating embedding...")
            embedding = self.ollama_processor.generate_embedding(description)
            
            # 5. Store in database
            print("   üíæ Storing in database...")
            image_id = self.db_helper.add_image(
                filename=image_path.name,
                file_path=str(image_path),
                file_size=metadata.get('file_size'),
                width=metadata.get('width'),
                height=metadata.get('height'),
                format=metadata.get('format'),
                description=description,
                tags=tags,
                embedding=embedding
            )
            
            if image_id:
                # 6. Add to FAISS index
                self.faiss_index.add(embedding.reshape(1, -1))
                current_index = self.faiss_index.ntotal - 1
                self.image_id_mapping[current_index] = image_id
                
                print(f"   ‚úÖ Successfully processed {image_path.name} (ID: {image_id})")
                return {
                    'image_id': image_id,
                    'filename': image_path.name,
                    'description': description,
                    'tags': tags,
                    'embedding_shape': embedding.shape,
                    'status': 'processed'
                }
            else:
                print(f"   ‚ùå Failed to store {image_path.name}")
                return None
                
        except Exception as e:
            print(f"‚ùå Error processing {image_path}: {e}")
            return None
    
    def process_images_folder(self, folder_path, max_images=None, save_every=10):
        """Process all images in a folder with automatic disk saves"""
        image_files = self.scan_images_folder(folder_path)
        
        if not image_files:
            print("‚ùå No images found to process")
            return []
        
        if max_images:
            image_files = image_files[:max_images]
            print(f"üî¢ Processing first {max_images} images")
        
        results = []
        processed_count = 0
        skipped_count = 0
        
        print(f"\nüöÄ Starting batch processing of {len(image_files)} images...")
        print(f"üíæ Auto-saving every {save_every} images")
        print("=" * 60)
        
        for i, image_path in enumerate(tqdm(image_files, desc="Processing images"), 1):
            print(f"\nüì∏ [{i}/{len(image_files)}] {image_path.name}")
            result = self.process_single_image(image_path)
            
            if result:
                results.append(result)
                if result.get('status') == 'processed':
                    processed_count += 1
                elif result.get('status') == 'skipped':
                    skipped_count += 1
                
                # Auto-save every N images
                if processed_count > 0 and processed_count % save_every == 0:
                    print(f"\nüíæ Auto-saving progress...")
                    self.save_index_to_disk()
            
            # Small delay to avoid overwhelming Ollama
            import time
            time.sleep(0.3)  # Reduced delay for batch processing
        
        # Final save
        print(f"\nüíæ Final save to disk...")
        self.save_index_to_disk()
        
        print(f"\n‚úÖ Batch processing complete!")
        print(f"üìä Results:")
        print(f"   üÜï Newly processed: {processed_count}")
        print(f"   ‚è≠Ô∏è  Skipped (already done): {skipped_count}")
        print(f"   üìÅ Total results: {len(results)}")
        print(f"   üîç FAISS index size: {self.faiss_index.ntotal} vectors")
        
        return results
    
    def search_images(self, query, top_k=5, confidence_threshold=0.3):
        """Search for images using natural language query"""
        try:
            print(f"üîç Searching for: '{query}'")
            
            # Generate embedding for the query
            query_embedding = self.ollama_processor.generate_embedding(query)
            
            # Search in FAISS index
            scores, indices = self.faiss_index.search(
                query_embedding.reshape(1, -1), 
                min(top_k, self.faiss_index.ntotal)
            )
            
            results = []
            for score, idx in zip(scores[0], indices[0]):
                if score >= confidence_threshold:
                    image_id = self.image_id_mapping.get(idx)
                    if image_id:
                        image_record = self.db_helper.get_image_by_id(image_id)
                        if image_record:
                            results.append({
                                'image_id': image_id,
                                'filename': image_record.filename,
                                'file_path': image_record.file_path,
                                'description': image_record.description,
                                'tags': image_record.tags,
                                'confidence': float(score),
                                'width': image_record.width,
                                'height': image_record.height
                            })
            
            print(f"‚úÖ Found {len(results)} matching images")
            return results
            
        except Exception as e:
            print(f"‚ùå Search failed: {e}")
            return []

# Initialize the disk-based pipeline
print("üîß Initializing disk-based pipeline...")
disk_pipeline = DiskBasedImageSearchPipeline(ollama_processor, db_helper)

print(f"\nüìä Current Status:")
print(f"   üóÑÔ∏è  Database: {len(db_helper.get_all_images())} images")
print(f"   üîç FAISS Index: {disk_pipeline.faiss_index.ntotal} vectors")
print(f"   üíæ Storage: Disk-based persistence enabled")

print(f"\nüéØ Ready for large-scale processing!")
print(f"üí° Usage for 100+ images:")
print(f"   results = disk_pipeline.process_images_folder(Config.IMAGES_FOLDER)")
print(f"   # Automatically saves every 10 images")
print(f"   # Skips already processed images")
print(f"   # Persistent across notebook restarts")
```

```python
# üîç SEARCH SYSTEM ANALYSIS & IMPROVEMENT
print("üîç ANALYZING AND IMPROVING SEARCH SYSTEM")
print("=" * 60)

def analyze_current_search_system():
    """Analyze the current search system to identify issues"""
    print("üìä CURRENT SEARCH SYSTEM ANALYSIS:")
    print("-" * 40)
    
    # 1. Check database embeddings
    conn = sqlite3.connect(Config.DATABASE_PATH)
    cursor = conn.cursor()
    
    cursor.execute("SELECT COUNT(*) FROM images WHERE embedding IS NOT NULL")
    with_embeddings = cursor.fetchone()[0]
    
    cursor.execute("SELECT COUNT(*) FROM images")
    total_images = cursor.fetchone()[0]
    
    print(f"üì∏ Images in database: {total_images}")
    print(f"üî¢ Images with embeddings: {with_embeddings}")
    
    # 2. Check a sample embedding
    if with_embeddings > 0:
        cursor.execute("SELECT embedding, description FROM images WHERE embedding IS NOT NULL LIMIT 1")
        sample = cursor.fetchone()
        if sample:
            embedding_json, description = sample
            embedding_data = json.loads(embedding_json)
            print(f"üìè Embedding dimension: {len(embedding_data)}")
            print(f"üìù Sample description: {description[:100]}...")
            print(f"üî¢ Sample embedding values: {embedding_data[:5]}...")
    
    # 3. Check FAISS index
    print(f"\nüîç FAISS INDEX STATUS:")
    print(f"   Vectors: {disk_pipeline.faiss_index.ntotal}")
    print(f"   Dimension: {disk_pipeline.faiss_index.d}")
    print(f"   Mappings: {len(disk_pipeline.image_id_mapping)}")
    
    conn.close()
    
    # 4. Test embedding generation
    print(f"\nüß™ TESTING EMBEDDING GENERATION:")
    test_queries = ["car", "sunset", "person", "building"]
    for query in test_queries:
        embedding = ollama_processor.generate_embedding(query)
        print(f"   '{query}' ‚Üí {embedding.shape}, norm: {np.linalg.norm(embedding):.3f}")

def test_search_with_debug(query, top_k=5):
    """Enhanced search with debugging information"""
    print(f"\nüîç DEBUG SEARCH: '{query}'")
    print("-" * 50)
    
    # 1. Generate query embedding
    query_embedding = ollama_processor.generate_embedding(query)
    print(f"üìù Query embedding: shape={query_embedding.shape}, norm={np.linalg.norm(query_embedding):.3f}")
    
    # 2. Search in FAISS
    if disk_pipeline.faiss_index.ntotal == 0:
        print("‚ùå FAISS index is empty!")
        return []
    
    search_k = min(top_k, disk_pipeline.faiss_index.ntotal)
    scores, indices = disk_pipeline.faiss_index.search(
        query_embedding.reshape(1, -1), 
        search_k
    )
    
    print(f"üîç FAISS search results:")
    print(f"   Scores: {scores[0]}")
    print(f"   Indices: {indices[0]}")
    
    # 3. Process results
    results = []
    for i, (score, idx) in enumerate(zip(scores[0], indices[0])):
        print(f"\nüì∏ Result {i+1}:")
        print(f"   Score: {score:.4f}")
        print(f"   FAISS index: {idx}")
        
        image_id = disk_pipeline.image_id_mapping.get(idx)
        print(f"   Database ID: {image_id}")
        
        if image_id:
            image_record = db_helper.get_image_by_id(image_id)
            if image_record:
                print(f"   Filename: {image_record.filename}")
                print(f"   Description: {image_record.description[:100]}...")
                
                # Calculate semantic similarity manually
                if image_record.embedding:
                    stored_embedding = np.array(json.loads(image_record.embedding))
                    similarity = np.dot(query_embedding, stored_embedding) / (
                        np.linalg.norm(query_embedding) * np.linalg.norm(stored_embedding)
                    )
                    print(f"   Manual similarity: {similarity:.4f}")
                
                results.append({
                    'image_id': image_id,
                    'filename': image_record.filename,
                    'description': image_record.description,
                    'confidence': float(score),
                    'faiss_index': idx
                })
    
    return results

def improved_search_system():
    """Create an improved search system with better relevance"""
    
    class ImprovedSearchPipeline:
        def __init__(self, base_pipeline):
            self.base_pipeline = base_pipeline
            
        def enhanced_search(self, query, top_k=5, confidence_threshold=0.1, use_description_search=True):
            """Enhanced search with multiple strategies"""
            print(f"üîç Enhanced search for: '{query}'")
            
            results = []
            
            # Strategy 1: Vector similarity search
            try:
                query_embedding = ollama_processor.generate_embedding(query)
                
                if self.base_pipeline.faiss_index.ntotal > 0:
                    scores, indices = self.base_pipeline.faiss_index.search(
                        query_embedding.reshape(1, -1), 
                        min(top_k * 2, self.base_pipeline.faiss_index.ntotal)  # Get more candidates
                    )
                    
                    for score, idx in zip(scores[0], indices[0]):
                        if score >= confidence_threshold:
                            image_id = self.base_pipeline.image_id_mapping.get(idx)
                            if image_id:
                                image_record = db_helper.get_image_by_id(image_id)
                                if image_record:
                                    results.append({
                                        'image_id': image_id,
                                        'filename': image_record.filename,
                                        'file_path': image_record.file_path,
                                        'description': image_record.description,
                                        'tags': image_record.tags,
                                        'confidence': float(score),
                                        'width': image_record.width,
                                        'height': image_record.height,
                                        'search_method': 'vector'
                                    })
            except Exception as e:
                print(f"‚ö†Ô∏è Vector search failed: {e}")
            
            # Strategy 2: Text-based description search (fallback/supplement)
            if use_description_search and len(results) < top_k:
                try:
                    text_results = db_helper.search_by_description(query)
                    for img in text_results[:top_k]:
                        # Avoid duplicates
                        if not any(r['image_id'] == img.id for r in results):
                            results.append({
                                'image_id': img.id,
                                'filename': img.filename,
                                'file_path': img.file_path,
                                'description': img.description,
                                'tags': img.tags,
                                'confidence': 0.8,  # Fixed confidence for text search
                                'width': img.width,
                                'height': img.height,
                                'search_method': 'text'
                            })
                except Exception as e:
                    print(f"‚ö†Ô∏è Text search failed: {e}")
            
            # Strategy 3: Tag-based search
            try:
                conn = sqlite3.connect(Config.DATABASE_PATH)
                cursor = conn.cursor()
                
                cursor.execute("""
                    SELECT id, filename, file_path, description, tags, width, height
                    FROM images 
                    WHERE tags LIKE ? 
                    LIMIT ?
                """, (f'%{query}%', top_k))
                
                tag_results = cursor.fetchall()
                conn.close()
                
                for row in tag_results:
                    img_id, filename, file_path, description, tags, width, height = row
                    # Avoid duplicates
                    if not any(r['image_id'] == img_id for r in results):
                        results.append({
                            'image_id': img_id,
                            'filename': filename,
                            'file_path': file_path,
                            'description': description,
                            'tags': tags,
                            'confidence': 0.7,  # Fixed confidence for tag search
                            'width': width,
                            'height': height,
                            'search_method': 'tags'
                        })
                        
            except Exception as e:
                print(f"‚ö†Ô∏è Tag search failed: {e}")
            
            # Sort by confidence and limit results
            results.sort(key=lambda x: x['confidence'], reverse=True)
            results = results[:top_k]
            
            print(f"‚úÖ Found {len(results)} results using multiple strategies")
            
            # Show search method breakdown
            methods = {}
            for r in results:
                method = r.get('search_method', 'unknown')
                methods[method] = methods.get(method, 0) + 1
            
            if methods:
                print(f"üìä Search methods used: {dict(methods)}")
            
            return results
        
        def display_enhanced_results(self, results, max_results=5):
            """Display search results with method indicators"""
            if not results:
                print("‚ùå No results found")
                return
            
            print(f"\nüéØ Enhanced Search Results ({len(results)} found):")
            print("=" * 80)
            
            for i, result in enumerate(results[:max_results], 1):
                method = result.get('search_method', 'unknown')
                method_emoji = {'vector': 'üî¢', 'text': 'üìù', 'tags': 'üè∑Ô∏è'}.get(method, '‚ùì')
                
                print(f"\nüì∏ Result {i}: {method_emoji} ({method})")
                print(f"   üìÅ File: {result['filename']}")
                print(f"   üìä Confidence: {result['confidence']:.3f}")
                print(f"   üìè Size: {result.get('width', '?')}x{result.get('height', '?')}")
                print(f"   üìù Description: {result['description'][:100]}...")
                if result.get('tags'):
                    print(f"   üè∑Ô∏è  Tags: {result['tags']}")
    
    return ImprovedSearchPipeline(disk_pipeline)

# Run analysis
analyze_current_search_system()

# Create improved search system
print(f"\nüöÄ CREATING IMPROVED SEARCH SYSTEM...")
improved_pipeline = improved_search_system()

# Test with debug search
print(f"\nüß™ TESTING DEBUG SEARCH...")
debug_results = test_search_with_debug("car", top_k=3)

print(f"\nüîß IMPROVED SEARCH SYSTEM READY!")
print(f"üí° New functions available:")
print(f"   ‚Ä¢ improved_pipeline.enhanced_search('query') - Multi-strategy search")
print(f"   ‚Ä¢ test_search_with_debug('query') - Debug search with details")
print(f"   ‚Ä¢ improved_pipeline.display_enhanced_results(results) - Better result display")
```

```python
# üåê CREATE STREAMLIT WEB UI FOR IMAGE SEARCH SYSTEM
print("üåê CREATING COMPREHENSIVE STREAMLIT WEB UI")
print("=" * 60)

# Create the complete Streamlit application
streamlit_app_code = '''
import streamlit as st
import sys
import os
from pathlib import Path
import sqlite3
import json
import time
from PIL import Image
import numpy as np
import pandas as pd
from datetime import datetime
import io
import hashlib

# Add project directory to path
PROJECT_ROOT = Path(__file__).parent
sys.path.append(str(PROJECT_ROOT))

# Import required libraries
try:
    import ollama
    import faiss
    from sqlalchemy import create_engine, Column, Integer, String, Text, Float, DateTime
    from sqlalchemy.ext.declarative import declarative_base
    from sqlalchemy.orm import sessionmaker
except ImportError as e:
    st.error(f"‚ùå Missing required packages: {e}")
    st.info("üí° Install with: pip install streamlit ollama faiss-cpu pillow numpy pandas sqlalchemy")
    st.stop()

# Page configuration
st.set_page_config(
    page_title="üîç AI Image Search System",
    page_icon="üñºÔ∏è",
    layout="wide",
    initial_sidebar_state="expanded"
)

# Custom CSS for better UI
st.markdown("""
<style>
    .main-header {
        background: linear-gradient(90deg, #667eea 0%, #764ba2 100%);
        padding: 2rem;
        border-radius: 10px;
        color: white;
        text-align: center;
        margin-bottom: 2rem;
    }
    .upload-section {
        border: 2px dashed #4CAF50;
        border-radius: 10px;
        padding: 2rem;
        text-align: center;
        background-color: #f9f9f9;
    }
    .search-result {
        border: 1px solid #ddd;
        border-radius: 8px;
        padding: 1rem;
        margin: 0.5rem 0;
        background: #f8f9fa;
    }
    .metric-card {
        background: white;
        padding: 1rem;
        border-radius: 8px;
        border-left: 4px solid #667eea;
        margin: 0.5rem 0;
    }
    .stProgress > div > div > div > div {
        background-color: #667eea;
    }
</style>
""", unsafe_allow_html=True)

# Configuration class
class Config:
    PROJECT_ROOT = Path(__file__).parent
    DATABASE_PATH = PROJECT_ROOT / "image_search.db"
    IMAGES_FOLDER = PROJECT_ROOT / "images"
    FAISS_INDEX_PATH = PROJECT_ROOT / "faiss_indexes"
    EMBEDDING_DIMENSION = 384
    DATABASE_URL = f"sqlite:///{DATABASE_PATH}"

# Ensure directories exist
Config.IMAGES_FOLDER.mkdir(exist_ok=True)
Config.FAISS_INDEX_PATH.mkdir(exist_ok=True)

# Database Models
Base = declarative_base()

class ImageRecord(Base):
    __tablename__ = 'images'
    id = Column(Integer, primary_key=True, autoincrement=True)
    filename = Column(String(255), nullable=False)
    file_path = Column(Text, nullable=False)
    file_size = Column(Integer)
    width = Column(Integer)
    height = Column(Integer)
    format = Column(String(10))
    created_at = Column(DateTime, default=datetime.now)
    processed_at = Column(DateTime)
    description = Column(Text)
    tags = Column(Text)
    embedding = Column(Text)

# Database initialization
@st.cache_resource
def init_database():
    try:
        engine = create_engine(Config.DATABASE_URL, echo=False)
        Base.metadata.create_all(engine)
        SessionLocal = sessionmaker(bind=engine)
        return engine, SessionLocal
    except Exception as e:
        st.error(f"Database initialization failed: {e}")
        return None, None

# Database helper class
class ImageDatabase:
    def __init__(self, session_factory):
        self.SessionLocal = session_factory
    
    def add_image(self, filename, file_path, file_size=None, width=None, height=None, 
                  format=None, description=None, tags=None, embedding=None):
        session = self.SessionLocal()
        try:
            # Convert embedding to JSON if needed
            embedding_json = None
            if embedding is not None:
                embedding_json = json.dumps(embedding.tolist())
            
            # Convert tags to string
            tags_str = None
            if tags:
                tags_str = ', '.join(tags) if isinstance(tags, list) else tags
            
            image_record = ImageRecord(
                filename=filename,
                file_path=file_path,
                file_size=file_size,
                width=width,
                height=height,
                format=format,
                description=description,
                tags=tags_str,
                embedding=embedding_json,
                processed_at=datetime.now()
            )
            
            session.add(image_record)
            session.commit()
            return image_record.id
        except Exception as e:
            session.rollback()
            st.error(f"Database error: {e}")
            return None
        finally:
            session.close()
    
    def get_all_images(self):
        session = self.SessionLocal()
        try:
            return session.query(ImageRecord).all()
        finally:
            session.close()
    
    def get_image_by_id(self, image_id):
        session = self.SessionLocal()
        try:
            return session.query(ImageRecord).filter(ImageRecord.id == image_id).first()
        finally:
            session.close()
    
    def search_by_description(self, search_term):
        session = self.SessionLocal()
        try:
            return session.query(ImageRecord).filter(
                ImageRecord.description.contains(search_term)
            ).all()
        finally:
            session.close()
    
    def check_duplicate(self, filename, file_size):
        session = self.SessionLocal()
        try:
            return session.query(ImageRecord).filter(
                ImageRecord.filename == filename,
                ImageRecord.file_size == file_size
            ).first()
        finally:
            session.close()

# Ollama processor
@st.cache_resource
def init_ollama_processor():
    class OllamaImageProcessor:
        def __init__(self):
            self.client = ollama
            self.vision_model = "llava"
            self.embedding_model = "nomic-embed-text"
            self.available_models = []
            self._check_ollama_connection()
        
        def _check_ollama_connection(self):
            try:
                models = ollama.list()
                self.available_models = [model['name'].split(':')[0] for model in models['models']]
                return True
            except Exception as e:
                st.error(f"‚ùå Ollama connection failed: {e}")
                st.info("üí° Make sure Ollama is running: ollama serve")
                return False
        
        def describe_image(self, image_path):
            try:
                response = ollama.chat(
                    model=self.vision_model,
                    messages=[{
                        'role': 'user',
                        'content': 'Describe this image in detail, including objects, people, colors, setting, mood, and any text visible.',
                        'images': [str(image_path)]
                    }]
                )
                return response['message']['content']
            except Exception as e:
                st.error(f"‚ùå Image description failed: {e}")
                return None
        
        def extract_image_tags(self, image_path):
            try:
                response = ollama.chat(
                    model=self.vision_model,
                    messages=[{
                        'role': 'user',
                        'content': 'List the main objects, people, and concepts in this image as short tags or keywords. Format as comma-separated tags like: sunset, mountains, orange sky, peaceful, landscape',
                        'images': [str(image_path)]
                    }]
                )
                tags_text = response['message']['content']
                tags = [tag.strip() for tag in tags_text.split(',')]
                return tags[:20]  # Limit to 20 tags
            except Exception as e:
                st.error(f"‚ùå Tag extraction failed: {e}")
                return []
        
        def generate_embedding(self, text):
            try:
                if self.embedding_model in self.available_models:
                    response = ollama.embeddings(model=self.embedding_model, prompt=text)
                    return np.array(response['embedding'])
                else:
                    return self._simple_text_embedding(text)
            except Exception as e:
                return self._simple_text_embedding(text)
        
        def _simple_text_embedding(self, text, dimension=384):
            words = text.lower().split()
            embedding = np.zeros(dimension)
            
            for word in words:
                hash1 = hash(word) % dimension
                hash2 = hash(word[::-1]) % dimension
                hash3 = hash(word + str(len(word))) % dimension
                
                embedding[hash1] += 1.0
                embedding[hash2] += 0.5
                embedding[hash3] += 0.3
            
            norm = np.linalg.norm(embedding)
            if norm > 0:
                embedding = embedding / norm
            
            return embedding
    
    return OllamaImageProcessor()

# FAISS search pipeline
@st.cache_resource
def init_search_pipeline():
    class StreamlitSearchPipeline:
        def __init__(self, ollama_processor, db_helper):
            self.ollama_processor = ollama_processor
            self.db_helper = db_helper
            self.faiss_index = None
            self.image_id_mapping = {}
            
            self.index_file = Config.FAISS_INDEX_PATH / "image_vectors.index"
            self.mapping_file = Config.FAISS_INDEX_PATH / "id_mapping.json"
            
            self._initialize_faiss_index()
        
        def _initialize_faiss_index(self):
            try:
                if self.index_file.exists() and self.mapping_file.exists():
                    self._load_index_from_disk()
                else:
                    self._create_new_index()
            except Exception as e:
                st.warning(f"FAISS initialization issue: {e}")
                self._create_new_index()
        
        def _create_new_index(self):
            self.faiss_index = faiss.IndexFlatIP(Config.EMBEDDING_DIMENSION)
            self.image_id_mapping = {}
        
        def _load_index_from_disk(self):
            try:
                self.faiss_index = faiss.read_index(str(self.index_file))
                with open(self.mapping_file, 'r') as f:
                    mapping_str = json.load(f)
                    self.image_id_mapping = {int(k): v for k, v in mapping_str.items()}
            except Exception as e:
                st.error(f"Failed to load FAISS index: {e}")
                self._create_new_index()
        
        def _save_index_to_disk(self):
            try:
                faiss.write_index(self.faiss_index, str(self.index_file))
                with open(self.mapping_file, 'w') as f:
                    mapping_str = {str(k): v for k, v in self.image_id_mapping.items()}
                    json.dump(mapping_str, f, indent=2)
                return True
            except Exception as e:
                st.error(f"Failed to save index: {e}")
                return False
        
        def process_uploaded_image(self, uploaded_file, progress_callback=None):
            """Process a single uploaded image"""
            try:
                # Generate unique filename
                timestamp = int(time.time())
                file_hash = hashlib.md5(uploaded_file.getvalue()).hexdigest()[:8]
                filename = f"{timestamp}_{file_hash}_{uploaded_file.name}"
                file_path = Config.IMAGES_FOLDER / filename
                
                # Check for duplicates
                file_size = len(uploaded_file.getvalue())
                existing = self.db_helper.check_duplicate(uploaded_file.name, file_size)
                if existing:
                    return None, "Image already exists in database"
                
                # Save the uploaded file
                with open(file_path, "wb") as f:
                    f.write(uploaded_file.getbuffer())
                
                if progress_callback:
                    progress_callback(0.2, "üìÅ Image saved")
                
                # Get image metadata
                with Image.open(file_path) as img:
                    width, height = img.size
                    format_type = img.format
                
                if progress_callback:
                    progress_callback(0.3, "üìè Metadata extracted")
                
                # Generate AI description
                description = self.ollama_processor.describe_image(file_path)
                if not description:
                    return None, "Failed to generate AI description"
                
                if progress_callback:
                    progress_callback(0.6, "üìù Description generated")
                
                # Extract tags
                tags = self.ollama_processor.extract_image_tags(file_path)
                
                if progress_callback:
                    progress_callback(0.8, "üè∑Ô∏è Tags extracted")
                
                # Generate embedding
                embedding = self.ollama_processor.generate_embedding(description)
                
                if progress_callback:
                    progress_callback(0.9, "üî¢ Embedding created")
                
                # Store in database
                image_id = self.db_helper.add_image(
                    filename=filename,
                    file_path=str(file_path),
                    file_size=file_size,
                    width=width,
                    height=height,
                    format=format_type,
                    description=description,
                    tags=tags,
                    embedding=embedding
                )
                
                if image_id:
                    # Add to FAISS index
                    self.faiss_index.add(embedding.reshape(1, -1))
                    current_index = self.faiss_index.ntotal - 1
                    self.image_id_mapping[current_index] = image_id
                    
                    # Save index
                    self._save_index_to_disk()
                    
                    if progress_callback:
                        progress_callback(1.0, "‚úÖ Processing complete")
                    
                    return image_id, {
                        'filename': filename,
                        'description': description,
                        'tags': tags,
                        'width': width,
                        'height': height,
                        'file_size': file_size,
                        'file_path': str(file_path)
                    }
                
                return None, "Failed to store in database"
                
            except Exception as e:
                return None, f"Processing error: {str(e)}"
        
        def search_images(self, query, top_k=5):
            """Multi-strategy image search"""
            results = []
            
            try:
                # Strategy 1: Vector similarity search
                if self.faiss_index.ntotal > 0:
                    query_embedding = self.ollama_processor.generate_embedding(query)
                    scores, indices = self.faiss_index.search(
                        query_embedding.reshape(1, -1), 
                        min(top_k * 2, self.faiss_index.ntotal)
                    )
                    
                    for score, idx in zip(scores[0], indices[0]):
                        if score >= 0.05:  # Low threshold for more results
                            image_id = self.image_id_mapping.get(idx)
                            if image_id:
                                image_record = self.db_helper.get_image_by_id(image_id)
                                if image_record:
                                    results.append({
                                        'image_id': image_id,
                                        'filename': image_record.filename,
                                        'file_path': image_record.file_path,
                                        'description': image_record.description,
                                        'tags': image_record.tags,
                                        'confidence': float(score),
                                        'width': image_record.width,
                                        'height': image_record.height,
                                        'method': 'vector'
                                    })
                
                # Strategy 2: Text search fallback
                if len(results) < top_k:
                    text_results = self.db_helper.search_by_description(query)
                    for img in text_results[:top_k]:
                        if not any(r['image_id'] == img.id for r in results):
                            results.append({
                                'image_id': img.id,
                                'filename': img.filename,
                                'file_path': img.file_path,
                                'description': img.description,
                                'tags': img.tags,
                                'confidence': 0.7,
                                'width': img.width,
                                'height': img.height,
                                'method': 'text'
                            })
                
                # Sort by confidence
                results.sort(key=lambda x: x['confidence'], reverse=True)
                return results[:top_k]
                
            except Exception as e:
                st.error(f"Search error: {e}")
                return []
    
    # Initialize components
    db_engine, SessionLocal = init_database()
    if not SessionLocal:
        st.error("Failed to initialize database")
        st.stop()
    
    db_helper = ImageDatabase(SessionLocal)
    ollama_processor = init_ollama_processor()
    
    return StreamlitSearchPipeline(ollama_processor, db_helper)

# Main application
def main():
    # Header
    st.markdown("""
    <div class="main-header">
        <h1>üîç AI Image Search System</h1>
        <p>Upload, process, and search images with AI-powered descriptions using LLaVA & FAISS</p>
    </div>
    """, unsafe_allow_html=True)
    
    # Initialize system
    try:
        pipeline = init_search_pipeline()
    except Exception as e:
        st.error(f"‚ùå System initialization failed: {e}")
        st.stop()
    
    # Sidebar with system status
    with st.sidebar:
        st.header("üìä System Status")
        
        try:
            all_images = pipeline.db_helper.get_all_images()
            total_images = len(all_images)
            vector_count = pipeline.faiss_index.ntotal if pipeline.faiss_index else 0
            
            col1, col2 = st.columns(2)
            with col1:
                st.metric("üñºÔ∏è Images", total_images)
            with col2:
                st.metric("üî¢ Vectors", vector_count)
            
            st.metric("üìè Embedding Dim", Config.EMBEDDING_DIMENSION)
            
            if total_images > 0:
                st.success("‚úÖ System Ready")
            else:
                st.info("üîÑ Upload images to start")
                
        except Exception as e:
            st.error(f"Status error: {e}")
        
        st.markdown("---")
        
        # Quick actions
        if st.button("üîÑ Refresh Status"):
            st.rerun()
        
        if st.button("üíæ Save Index"):
            if pipeline._save_index_to_disk():
                st.success("üíæ Index saved!")
            else:
                st.error("‚ùå Save failed!")
        
        # Ollama status
        st.markdown("---")
        st.subheader("ü§ñ Ollama Status")
        try:
            models = pipeline.ollama_processor.available_models
            if "llava" in models:
                st.success("‚úÖ LLaVA available")
            else:
                st.warning("‚ö†Ô∏è LLaVA missing")
        except:
            st.error("‚ùå Ollama not connected")
    
    # Main content tabs
    tab1, tab2, tab3 = st.tabs(["üì§ Upload Images", "üîç Search Images", "üìã Browse Database"])
    
    # Upload Tab
    with tab1:
        st.header("üì§ Upload & Process Images")
        
        # File uploader
        uploaded_files = st.file_uploader(
            "Choose images to upload and process",
            type=['png', 'jpg', 'jpeg', 'bmp', 'tiff', 'webp'],
            accept_multiple_files=True,
            help="Select one or more images. They will be automatically processed with AI descriptions and added to the searchable database."
        )
        
        if uploaded_files:
            st.success(f"üìÅ {len(uploaded_files)} file(s) selected")
            
            # Processing options
            col1, col2 = st.columns([2, 1])
            
            with col1:
                process_button = st.button("üöÄ Process All Images", type="primary", use_container_width=True)
            
            with col2:
                show_details = st.checkbox("Show detailed progress", value=True)
            
            if process_button:
                st.markdown("---")
                st.subheader("üîÑ Processing Images...")
                
                # Overall progress
                overall_progress = st.progress(0)
                overall_status = st.empty()
                
                # Results container
                results_container = st.container()
                
                successful_uploads = 0
                failed_uploads = 0
                
                for i, uploaded_file in enumerate(uploaded_files):
                    # Update overall progress
                    progress = i / len(uploaded_files)
                    overall_progress.progress(progress)
                    overall_status.text(f"Processing {i+1}/{len(uploaded_files)}: {uploaded_file.name}")
                    
                    with results_container:
                        if show_details:
                            st.write(f"**üñºÔ∏è Processing: {uploaded_file.name}**")
                            
                            # Individual progress
                            detail_progress = st.progress(0)
                            detail_status = st.empty()
                            
                            def progress_callback(progress_val, message):
                                detail_progress.progress(progress_val)
                                detail_status.text(message)
                        else:
                            progress_callback = None
                        
                        # Process the image
                        image_id, result = pipeline.process_uploaded_image(uploaded_file, progress_callback)
                        
                        if image_id:
                            successful_uploads += 1
                            if show_details:
                                st.success(f"‚úÖ Successfully processed: {uploaded_file.name}")
                                
                                # Show image preview and details
                                col1, col2 = st.columns([1, 2])
                                
                                with col1:
                                    try:
                                        img_path = Path(result['file_path'])
                                        if img_path.exists():
                                            st.image(str(img_path), width=200, caption=result['filename'])
                                    except:
                                        st.write("üñºÔ∏è Preview unavailable")
                                
                                with col2:
                                    st.write(f"**üìÅ Filename:** {result['filename']}")
                                    st.write(f"**üìè Dimensions:** {result['width']}√ó{result['height']}")
                                    st.write(f"**üíæ Size:** {result['file_size']/1024:.1f} KB")
                                    st.write(f"**üìù Description:** {result['description'][:150]}...")
                                    if result['tags']:
                                        st.write(f"**üè∑Ô∏è Tags:** {', '.join(result['tags'][:8])}")
                            else:
                                st.success(f"‚úÖ {uploaded_file.name}")
                        else:
                            failed_uploads += 1
                            if show_details:
                                st.error(f"‚ùå Failed to process: {uploaded_file.name}")
                                st.write(f"**Error:** {result}")
                            else:
                                st.error(f"‚ùå {uploaded_file.name}: {result}")
                        
                        if show_details:
                            st.markdown("---")
                
                # Final summary
                overall_progress.progress(1.0)
                overall_status.text("‚úÖ Processing complete!")
                
                st.success(f"üéâ Processing complete! ‚úÖ {successful_uploads} successful, ‚ùå {failed_uploads} failed")
                
                if successful_uploads > 0:
                    st.balloons()
    
    # Search Tab
    with tab2:
        st.header("üîç Search Images")
        
        # Search interface
        col1, col2 = st.columns([4, 1])
        
        with col1:
            search_query = st.text_input(
                "What are you looking for?",
                placeholder="e.g., sunset, car, person walking, red building, technology...",
                help="Use natural language to describe what you want to find"
            )
        
        with col2:
            num_results = st.selectbox("Results", [5, 10, 15, 20], index=0)
        
        # Search button and results
        if search_query:
            if st.button("üîç Search", type="primary"):
                with st.spinner(f"üîç Searching for '{search_query}'..."):
                    results = pipeline.search_images(search_query, top_k=num_results)
                
                if results:
                    st.success(f"üéØ Found {len(results)} matching images")
                    
                    # Display results
                    for i, result in enumerate(results, 1):
                        with st.expander(f"üì∏ Result {i}: {result['filename']}", expanded=True):
                            col1, col2 = st.columns([1, 2])
                            
                            with col1:
                                # Display image
                                try:
                                    img_path = Path(result['file_path'])
                                    if img_path.exists():
                                        st.image(str(img_path), width=300)
                                    else:
                                        st.warning("üñºÔ∏è Image file not found")
                                except Exception as e:
                                    st.error(f"Image display error: {e}")
                            
                            with col2:
                                # Display metadata
                                method_emoji = {'vector': 'üî¢', 'text': 'üìù', 'tags': 'üè∑Ô∏è'}.get(result.get('method'), '‚ùì')
                                
                                st.write(f"**Search Method:** {method_emoji} {result.get('method', 'unknown').title()}")
                                st.write(f"**Confidence:** {result['confidence']:.3f}")
                                st.write(f"**Dimensions:** {result.get('width', '?')}√ó{result.get('height', '?')}")
                                
                                st.write("**AI Description:**")
                                st.write(result['description'])
                                
                                if result.get('tags'):
                                    st.write("**Tags:**")
                                    st.write(result['tags'])
                else:
                    st.info("üîç No matching images found. Try different search terms or upload more images.")
    
    # Browse Tab
    with tab3:
        st.header("üìã Database Browser")
        
        try:
            all_images = pipeline.db_helper.get_all_images()
            
            if all_images:
                st.write(f"üìä **Total images in database:** {len(all_images)}")
                
                # Create summary table
                summary_data = []
                for img in all_images:
                    summary_data.append({
                        'ID': img.id,
                        'Filename': img.filename,
                        'Size': f"{img.width}√ó{img.height}" if img.width and img.height else "Unknown",
                        'Format': img.format or "Unknown",
                        'Processed': img.processed_at.strftime("%Y-%m-%d %H:%M") if img.processed_at else "Unknown",
                        'Description Preview': (img.description[:80] + "...") if img.description and len(img.description) > 80 else (img.description or "No description")
                    })
                
                # Display table
                df = pd.DataFrame(summary_data)
                st.dataframe(df, use_container_width=True, hide_index=True)
                
                st.markdown("---")
                
                # Image detail viewer
                st.subheader("üîç Image Details")
                
                selected_id = st.selectbox(
                    "Select an image to view details:",
                    options=[img.id for img in all_images],
                    format_func=lambda x: f"ID {x}: {next(img.filename for img in all_images if img.id == x)}"
                )
                
                if selected_id:
                    selected_img = pipeline.db_helper.get_image_by_id(selected_id)
                    
                    if selected_img:
                        col1, col2 = st.columns([1, 2])
                        
                        with col1:
                            # Display image
                            try:
                                img_path = Path(selected_img.file_path)
                                if img_path.exists():
                                    st.image(str(img_path), width=350)
                                else:
                                    st.warning("üñºÔ∏è Image file not found on disk")
                            except Exception as e:
                                st.error(f"Image display error: {e}")
                        
                        with col2:
                            # Display all details
                            st.write(f"**üìÅ Filename:** {selected_img.filename}")
                            st.write(f"**üÜî Database ID:** {selected_img.id}")
                            st.write(f"**üìè Dimensions:** {selected_img.width}√ó{selected_img.height}")
                            st.write(f"**üìä Format:** {selected_img.format}")
                            st.write(f"**üíæ File Size:** {selected_img.file_size/1024:.1f} KB" if selected_img.file_size else "Unknown")
                            st.write(f"**üìÖ Processed:** {selected_img.processed_at}")
                            st.write(f"**üìç File Path:** {selected_img.file_path}")
                            
                            st.write("**üìù AI Generated Description:**")
                            st.write(selected_img.description or "No description available")
                            
                            if selected_img.tags:
                                st.write("**üè∑Ô∏è Extracted Tags:**")
                                st.write(selected_img.tags)
                            
                            # Show embedding info
                            if selected_img.embedding:
                                embedding_data = json.loads(selected_img.embedding)
                                st.write(f"**üî¢ Embedding:** {len(embedding_data)}-dimensional vector")
            else:
                st.info("üì≠ No images in database yet. Upload some images to get started!")
                
        except Exception as e:
            st.error(f"Database browser error: {e}")

if __name__ == "__main__":
    main()
'''

# Save the Streamlit app
app_file = Config.PROJECT_ROOT / "streamlit_app.py"
with open(app_file, 'w', encoding='utf-8') as f:
    f.write(streamlit_app_code)

print(f"‚úÖ Streamlit app created: {app_file}")

# Create requirements.txt
requirements_content = """streamlit>=1.28.0
ollama>=0.1.7
faiss-cpu>=1.7.4
pillow>=10.0.0
numpy>=1.24.0
pandas>=2.0.0
sqlalchemy>=2.0.0
python-dotenv>=1.0.0
"""

requirements_file = Config.PROJECT_ROOT / "requirements.txt"
with open(requirements_file, 'w') as f:
    f.write(requirements_content)

print(f"‚úÖ Requirements file created: {requirements_file}")

# Create a simple run script
run_script_content = """#!/bin/bash
echo "üöÄ Starting AI Image Search System..."
echo "üìã Make sure Ollama is running: ollama serve"
echo "üîç Starting Streamlit app..."
streamlit run streamlit_app.py
"""

run_script_file = Config.PROJECT_ROOT / "run_app.sh"
with open(run_script_file, 'w') as f:
    f.write(run_script_content)

# Make script executable
import stat
run_script_file.chmod(stat.S_IRWXU | stat.S_IRGRP | stat.S_IROTH)

print(f"‚úÖ Run script created: {run_script_file}")

print(f"\nüéâ STREAMLIT WEB UI CREATED SUCCESSFULLY!")
print(f"=" * 60)

print(f"\nüìÅ Files Created:")
print(f"   üåê {app_file.name} - Main Streamlit application")
print(f"   üì¶ {requirements_file.name} - Python dependencies")
print(f"   üöÄ {run_script_file.name} - Easy launch script")

print(f"\nüí° How to Run:")
print(f"   1Ô∏è‚É£ Make sure Ollama is running:")
print(f"      ollama serve")
print(f"   ")
print(f"   2Ô∏è‚É£ Install dependencies (if needed):")
print(f"      pip install -r requirements.txt")
print(f"   ")
print(f"   3Ô∏è‚É£ Launch the app:")
print(f"      streamlit run streamlit_app.py")
print(f"      # OR simply: ./run_app.sh")
print(f"   ")
print(f"   4Ô∏è‚É£ Open browser:")
print(f"      http://localhost:8501")

print(f"\nüéØ Features Included:")
print(f"   üì§ Multi-file image upload with drag & drop")
print(f"   ü§ñ Real-time AI processing (LLaVA descriptions + tags)")
print(f"   üîç Smart multi-strategy search (vector + text + tags)")
print(f"   üìä Live system status dashboard")
print(f"   üìã Complete database browser")
print(f"   üíæ Automatic FAISS index persistence")
print(f"   üé® Beautiful responsive web interface")
print(f"   üîÑ Progress tracking for uploads")
print(f"   üéØ Confidence scoring for search results")
print(f"   üñºÔ∏è Image previews and detailed metadata")

print(f"\n‚ö†Ô∏è  Prerequisites:")
print(f"   ü¶ô Ollama running with LLaVA model")
print(f"   üì¶ Python packages installed")
print(f"   üìÅ Write permissions in project directory")

print(f"\nüöÄ Ready to launch your AI Image Search web app!")
```
